```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Read the data from the CSV file into a DataFrame and display the first five rows 

```python
# Read file root dir 'avocado.csv'
file = 'avocado.csv'
data_frame = pd.read_csv(file)
data_frame.head()
```

2. Display basic information about the DataFrame and its columns to see that the Date column has the object type 

```python
print(data_frame.dtypes)
# Date column has the object type.
```

3. Convert the Date column to the datetime type

```python
data_frame['Date'] = pd.to_datetime(data_frame['Date'])
```

4. Filter the data so it contains just the rows for 2015, for the conventional type, and for the region named Albany. Store just the Date, Total Bags, and Small Bags columns in a new DataFrame

```python
f_data_frame = data_frame[ (data_frame['year'] == 2015) 
    & (data_frame['type'] == 'conventional') 
    & (data_frame['region'] == 'Albany') ][['Date', 'Total Bags', 'Small Bags']].copy()
```

5. Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column.
   Name the column that contains the type of bag Bags, and name the column that contains the number of bags Count.
   Then, display the resulting new DataFrame

```python
m_data_frame = f_data_frame.melt(id_vars='Date',
                                 value_vars=['Total Bags', 'Small Bags'],
                                 var_name='Bags', value_name='Count')
```

6. Plot the melted data with Seaborn in a line plot, using the hue parameter to distinguish between the bag types

```python
plt.figure()
sns.lineplot(data=m_data_frame, x='Date', y='Count', hue='Bags')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend(title='Bags')
plt.show()
```

7. Bin the data in the Total Volume column into four quantiles labeled poor, modest, good, and excellent, and store the bin labels in a new column

```python
data_frame['Volume Category'] = pd.qcut(data_frame['Total Volume'], q=4, labels=['poor', 'modest', 'good', 'excellent'])
```

8. Plot the binned data by year using a Seaborn count plot

```python
plt.figure()
sns.countplot(data=data_frame, x='Volume Category', hue='year')
plt.xlabel('Volume Category')
plt.ylabel('Count')
plt.legend(title='Year')
plt.show()
```

Question 1. Of the three bag sizes, which size sells the most? Hint: Melt the three bag size columns and plot the bags by year

```python
m_count_data_frame = pd.melt(data_frame, 
                             id_vars=['year'],
                             value_vars=['Total Bags', 'Small Bags', 'Large Bags', 'XLarge Bags'],
                             var_name='Bag Size',
                             value_name='Count')

plt.figure()
sns.lineplot(data=m_count_data_frame, x='year', y='Count', hue='Bag Size')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Bag Size')
plt.tight_layout()
plt.show()

# ANSWER: Small Bags, as seen in the orange in the graph. 
```

Question 2: Which type of avocado sells the most

```python
sales_data_frame = data_frame.groupby('type')['Total Volume'].sum()

highest_sales = sales_data_frame.idxmax()
print(f'The highest sales by type is: {highest_sales}')
# Answer: Conventional
```
