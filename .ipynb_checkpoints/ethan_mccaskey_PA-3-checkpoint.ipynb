```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
```

1. Read the data from the .csv file into a DataFrame

```python
# Read csv file.
file = "avocado.csv"
# Reading the data
data_frame = pd.read_csv(file)
```

1-1. Display the first five rows

```python
# First five rows
first_five_rows = data_frame.head()
first_five_rows
```

2. Create a new DataFrame that contains the total of the Small Bags, Large Bags, and XLarge Bags columns grouped by type, and then display the DataFrame

```python
# Sum of different bag types. 
grouped_data_frame = data_frame.groupby('type')[['Small Bags', 'Large Bags', 'XLarge Bags']].sum().reset_index()
grouped_data_frame
```

3. Use the grouped data to create a bar plot that shows the number of small, large, and extra-large bags for both types of avocado

```python
grouped_data_frame.set_index('type', inplace=True)
grouped_data_frame.plot(kind='bar')

plot.title('Total Bags for each size and type')
plot.ylabel('Total Bags')
```

4. Use the original data to create a scatter plot for the Total Volume and AveragePrice columns

```python
plot.figure(figsize=(20, 6))
# Create scatter plot of total volume and average price. 
plot.scatter(data_frame['Total Volume'], data_frame['AveragePrice'])
plot.title('Total Volume and AveragePrice')
```

1. In what year were the most bags of avocados sold

```python
# Group by total bags and sum.
year_total_bags = data_frame.groupby('year')['Total Bags'].sum()
# Get the max return the index.
year_with_most_bags = year_total_bags.idxmax()

# Answer: 2017
year_with_most_bags
```

2. Do the total bags of avocados appear to have an effect on the average price

```python
# Get the correlation 
correlation = data_frame['Total Bags'].corr(data_frame['AveragePrice'])
# Answer: Because the correlation is near zero, there is a weak or almost no relationship between the variables.
correlation
```
