8. Multiply the Total Volume and AveragePrice columns, and store the result in a new column called EstimatedRevenue. Then, display the first five rows of this data to confirm that the column was added and has the correct values

```python
# New column 'EstimatedRevenue' (Total Volume * Average Price)
data_frame.EstimatedRevenue = data_frame['Total Volume'] * data_frame.AveragePrice
# Display first five rows
data_frame.head()
```

9. Create a DataFrame that’s grouped by region and type and that includes the average price for the grouped columns. Then, reset the index and display the first five row

```python
# DataFrame that is grouped by region and type, and has the AveragePrice.
grouped_data_frame = data_frame.groupby(['region', 'type']).AveragePrice
# Then calculate the average of AveragePrice inside of each group for unique
# Reset index puts them back to normal columns.
grouped_data_frame = grouped_data_frame.mean().reset_index()

# First five rows.
grouped_data_frame.head()
```

```python
10. Create a bar plot that shows the mean, median, and standard deviation of the Total Volume column by year 
```

```python
# Group data by year, then calculate the mean, median, and standard deviation of the Total Volume column by year.
graph_data = data_frame.groupby('year')['Total Volume'].agg(['mean', 'median', 'std'])
# Bar plot of data using Matplotlib
graph_data.plot(kind='bar', figsize=(10, 4), width=1.0, yerr=graph_data['std'], capsize=5)
```

2. What is the average price for each type of avocado (organic and conventional)? 

```python
# Answer: Conventional = 1.158040 | Organic = 1.653999
average_price_for_type = data_frame.groupby('type').AveragePrice.mean()
average_price_for_type
```

3. Which region has the lowest average price for organic avocados? Hint: Create wide data from the grouped data that you created in task 9

```python
# Answer: Houston (conventional = 0.825089 | organic = 1.270769)
# Wide data so region is index and type is column
wide_data = grouped_data_frame.pivot(index='region', columns='type', values='AveragePrice')
# Find the lowest
low_price_reg = wide_data[wide_data.organic == wide_data.organic.min()]
low_price_reg
```

Have the Total Bags sold per year of each type of avocado become more or less consistent over time 

```python
# Answer: Less consistent... I think.

# Group by year anad type and find std 
std_bags_yearly = data_frame.groupby(['year', 'type'])['Total Bags'].std().reset_index()
# Reshape it
std_bags_wide = std_bags_yearly.pivot(index='year', columns='type', values='Total Bags')
# I tried to reformat to percentage make more sense
std_bags_wide = std_bags_wide.apply(lambda x: x / x.mean() * 100)
# quick bar graph
std_bags_wide.plot(kind='bar', figsize=(10, 6))
std_bags_wide
```
