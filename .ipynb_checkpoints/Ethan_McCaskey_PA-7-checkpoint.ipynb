```python
import pandas as pd
```

1. Read the data from the .csv file into a DataFrame and display the first five rows

```python
# Read file root dir 'avocado.csv'
file = 'avocado.csv'
data_frame = pd.read_csv(file)
data_frame.head()
```

2. Notice that the column names are inconsistent. Change the column names so they all use title case with no spaces

```python
# Lambda func to do the operations on, which is making first character upper case, and removing spaces.
data_frame.rename(columns=lambda x: x.title().replace(" ", ""), inplace=True)
```

3. Add a column with the percentage of the bags in the TotalBags column that are extra-large bags

```python
data_frame['Percentageofxlargebags'] = (data_frame['Xlargebags'] / data_frame['Totalbags']) * 100
```

4. Add a column with the percentage of the bags in the TotalBags column that are large bags

```python
data_frame['Percentageoflargebags'] = (data_frame['Largebags'] / data_frame['Totalbags']) * 100
```

5. Add a column with the percentage of the bags in the TotalBags column that are small bags

```python
data_frame['Percentageofsmallbags'] = (data_frame['Smallbags'] / data_frame['Totalbags']) * 100
```

6. Display the first five rows of data to view the new columns

```python
data_frame.head()
```

7. Assign the Region, Type, Year, and TotalBags columns to a new DataFrame

```python
new_data_frame = data_frame[['Region', 'Type', 'Year', 'Totalbags']]
```

8. Group, sum, and unstack this data to produce a new DataFrame that looks like this

```python
grouped = data_frame.pivot_table(index=['Region', 'Year'], columns='Type', values='Totalbags', aggfunc='sum')
grouped
```
